```python
# yêu cầu: tìm dãy con liên tiếp có tổng lớn nhất
n = int(input())
arr = [int(i) for i in input().split()]

def finding_max_sum(arr, n):
  curr_max = p = q = tmp = 0
  global_max = arr[0]

  for i in range(0, len(arr)):
    curr_max += arr[i]
    if curr_max < 0:
       curr_max = 0 
       tmp = i + 1

    if global_max < curr_max:
      global_max = curr_max
      p = tmp + 1
      q = i + 1 

  return p, q, global_max

p, q, sum = finding_max_sum(arr, n)
print(p, q, sum)
```

# 1) Abstraction:


Cho số nguyên n và mảng một chiều gồm n phần tử, tìm tổng lớn nhất có thể có của một mảng con liền kề trong mảng một chiều đó

# 2) Pattern Recognition:

Tên dạng bài toán: Tìm mảng con có tổng lớn nhất

Phương pháp: Thuật toán Kadane

Ý tưởng giải thuật áp dụng cho bài toán:
+ Duyệt mảng từ trái sang phải
+ Với mỗi dãy con tính từ vị trí đầu đến vị trí hiện tại, xác định tổng lớn nhất có thể có của dãy con hiện tại
+ Các tổng lớn nhất của từng dãy con được mang ra so sánh với nhau để tìm ra tổng lớn nhất. Cụ thể hóa trong bài toán minh họa trên bằng việc chọn ra biến global_sum để chứa tổng lớn nhất của cả quá trình
