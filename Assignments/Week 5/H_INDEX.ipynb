```python
from sys import stdin
import bisect
n = int(stdin.readline())

if(n > 0):
    arr = [int(x) for x in stdin.readline().split(" ",n-1)]
    arr.sort(reverse=False)
    while(n > 0):
        temp = bisect.bisect_right(arr,n)
        count = len(arr) - temp
        if(count < n):
            n -= 1
        else:
            break   
print(count)
```

# 1) Abstraction:

Nhập vào số n và giá trị của từng số trong dãy n số. 
Ta xét bài toán thỏa mãn điều kiện: "Tổng số lượng số trong dãy n số (giá trị của số tại từng vị trí[0,1,..,n] >= n) >= n"

 Nếu thỏa, xuất n

 Nếu không thỏa, n - 1 (n > 0)
