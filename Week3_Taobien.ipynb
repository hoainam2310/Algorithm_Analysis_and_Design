```python
#Tạo function dãy fibonacci
def fibonacci(value):
    p1 = 1
    p2 = 1
    for i in range(1, value):
        p3 = p1 + p2
        p2 = p1
        p1 = p3
    return p3

#nhập n và k trên cùng 1 dòng
n , k = map(int, input().split())

#Điều kiện của n
if((n >= 1) and (n <= 1000)):
    o = (n * fibonacci(2*k)) % 1000000007
    
print(o)
```

# 1) Abstraction:

Cho hai số nguyên n và k, tính n * fibonacci(2 * k) % (pow(10,9) + 7)

# 2) Pattern Recognition:

Tên dạng bài toán: Tìm số fibonacci thứ n

Kỹ thuật áp dụng: Dùng vòng lặp
