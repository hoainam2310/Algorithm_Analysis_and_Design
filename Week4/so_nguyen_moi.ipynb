```python
from sys import stdin
#Nhập chữ số n không quá 100 chữ số
n = int(stdin.readline())

temp = n
count = 0
arr = []
frag = False

#Đếm số lượng chữ số của n
while(temp > 0):
    arr.append(temp % 10)
    temp = temp // 10
    count += 1
#Đảo ngược lại mảng arr
arr = list(reversed(arr))

#Tổng các chữ số trong số n 
sum = 0
for i in range(count):
    sum = sum + arr[i]

#Xét từ số đầu tiên trong mảng arr
for i in range(count):  
    #Cho x chạy từ 9 trở về trước
    for x in range(9, arr[i] , -1):
        #Điều kiện chia hết cho 3
        if ((x + sum - arr[i]) % 3 == 0):
            print(n - arr[i]*(10**(count-1)) + x*(10**(count-1)))
            frag = True
            break
    if(frag == True):
        break
    else:
        count -= 1

if(count == 0):
    for x in range(arr[-1] - 1, 1, -1):
        if((x + sum - arr[-1]) % 3 == 0):
            print(n - arr[-1] + x)
            break
```

1) Abstraction:
Cho số nguyên dương n, tìm số m lớn nhất chia hết cho 3 và khác n một chữ số

2) Pattern Recognition:

+ Tên dạng bài toán: chuyển số số thực  

+ Kỹ thuật áp dụng: dùng Brute Force để vét cạn hết các trường hợp có trong mảng chứa các chữ số của n
