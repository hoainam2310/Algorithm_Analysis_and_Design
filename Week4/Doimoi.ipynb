```python
a, k, b, m, n = list(map(int, input().split()))
tmp = 0

#Số cây mà cả 2 đội cùng đốn hạ 
def Renewed(x):
    return a * (x - (x // k)) + b * (x - (x // m)) 

res = int(n // (a * (1 - 1 / k) + b * (1 - 1 / m))) - 1

#Xét điều kiện, duyệt tất cả các trường hợp từ x-1 đến x+1
while (tmp < n):
    tmp = Renewed(res)
    res += 1

print(res - 1)
```

# **1) Abstraction:**

Cho 5 số nguyên a, k, b, m, n, tìm giá trị nhỏ nhất của x sao cho x >= n / (((1 - 1/k) * a) + (1 - 1/m) * b)))


# **2) Pattern Recognition:**

Ý tưởng áp dụng cho bài toán:

Gọi x là số ngày 2 đội chặt hết n cây

  Đội 1
- Số ngày nghỉ: int(x/k)
- Số ngày làm: x - int(x/k)
- Số cây mà đội đốn hạ: a * (x - int(x/k))

  Đội 2
- Số ngày nghỉ: int(x/k)
- Số ngày làm: x - int(x/k)
- Số cây mà đội đốn hạ: b * (x - int(x/k))

=> Tổng số cây mà cả hai đội cùng đốn hạ: 

  a * (x - int(x/k)) + b * (x - int(x/k)) >= n

<=> x((1 - 1/k) * a + (1 - 1/m) * b) * b) >= n 

<=> x >= n // (a * (x - (x // k)) + b * (x - (x // m))

Lưu ý: Trong quá trình xét điều kiện, cụ thể là cần xét trong đoạn từ [x - 1; x + 1] do tồn tại sai số khi thực hiện phép chia.
